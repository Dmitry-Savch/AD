#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import numpy as np
import pandas as pd
from scipy.integrate import solve_ivp
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
def lorenz_system(t, state, sigma, rho, beta):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]

# Parameters for the Lorenz system
sigma = 10
rho = 28
beta = 8 / 3

# Time span and initial conditions
time_span = (0, 50)  # Simulate for 50 time units
initial_state = [1.0, 1.0, 1.0]
time_points = np.linspace(time_span[0], time_span[1], 10000)

# Solve the Lorenz system
solution = solve_ivp(lorenz_system, time_span, initial_state, t_eval=time_points, args=(sigma, rho, beta))
lorenz_data = pd.DataFrame({"t": solution.t, "x": solution.y[0], "y": solution.y[1], "z": solution.y[2]})
lorenz_data.to_csv("lorenz_data.csv", index=False)
```

```python
def generate_narendra_parthasarathy_data(alpha, beta, gamma, delta, num_points):
    np.random.seed(0)  # For reproducibility
    x1, x2, x3 = np.random.rand(3)  # Initial conditions
    data = []

    for _ in range(num_points):
        x1_next = alpha[0] * x1 + beta[0] * x2 + gamma[0] * x3 + delta[0]
        x2_next = alpha[1] * x1 + beta[1] * x2 + gamma[1] * x3 + delta[1]
        x3_next = alpha[2] * x1 + beta[2] * x2 + gamma[2] * x3 + delta[2]
        data.append([x1_next, x2_next, x3_next])
        x1, x2, x3 = x1_next, x2_next, x3_next

    return pd.DataFrame(data, columns=["x1", "x2", "x3"])

# Coefficients for Narendra-Parthasarathy model
alpha = [0.5, -0.3, 0.2]
beta = [0.1, 0.4, -0.2]
gamma = [-0.2, 0.1, 0.3]
delta = [0.05, -0.05, 0.1]

# Generate data
narendra_data = generate_narendra_parthasarathy_data(alpha, beta, gamma, delta, 10000)
narendra_data.to_csv("narendra.csv", index=False)
```

```python
lorenz_features = lorenz_data["x"]
lorenz_train, lorenz_temp = train_test_split(lorenz_features, test_size=0.3, random_state=42)
lorenz_val, lorenz_test = train_test_split(lorenz_temp, test_size=0.5, random_state=42)

# For Narendra-Parthasarathy data
narendra_train, narendra_temp = train_test_split(narendra_data, test_size=0.3, random_state=42)
narendra_val, narendra_test = train_test_split(narendra_temp, test_size=0.5, random_state=42)
```

```python
scaler_lorenz = StandardScaler()
scaler_narendra = StandardScaler()

# Normalize Lorenz data
lorenz_train_scaled = scaler_lorenz.fit_transform(lorenz_train.values.reshape(-1, 1))
lorenz_val_scaled = scaler_lorenz.transform(lorenz_val.values.reshape(-1, 1))
lorenz_test_scaled = scaler_lorenz.transform(lorenz_test.values.reshape(-1, 1))

# Normalize Narendra-Parthasarathy data
narendra_train_scaled = scaler_narendra.fit_transform(narendra_train)
narendra_val_scaled = scaler_narendra.transform(narendra_val)
narendra_test_scaled = scaler_narendra.transform(narendra_test)
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "SVM": SVR()
}

print("Training models and evaluating performance on validation data:")
model_results = []
for name, model in models.items():
    model.fit(lorenz_train_scaled, lorenz_train.values)
    val_predictions = model.predict(lorenz_val_scaled)
    test_predictions = model.predict(lorenz_test_scaled)

    mse_val = mean_squared_error(lorenz_val.values, val_predictions)
    mse_test = mean_squared_error(lorenz_test.values, test_predictions)
    r2_val = r2_score(lorenz_val.values, val_predictions)
    r2_test = r2_score(lorenz_test.values, test_predictions)

    model_results.append({
        "Model": name,
        "Validation MSE": mse_val,
        "Test MSE": mse_test,
        "Validation R2": r2_val,
        "Test R2": r2_test
    })

    print(f"{name} - Validation MSE: {mse_val:.4f}, Validation R2: {r2_val:.4f}, Test MSE: {mse_test:.4f}, Test R2: {r2_test:.4f}")
```

```python
results_df = pd.DataFrame(model_results)
results_df.to_csv("model_results.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lorenz_test.values, label="Actual Values", linestyle="--")
for name, model in models.items():
    test_predictions = model.predict(lorenz_test_scaled)
    plt.plot(test_predictions, label=f"{name} Predictions")
plt.legend()
plt.title("Model Predictions vs Actual Values")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for name, model in models.items():
    test_predictions = model.predict(lorenz_test_scaled)
    residuals = lorenz_test.values - test_predictions
    plt.plot(residuals, label=f"{name} Residuals")
plt.legend()
plt.title("Residual Analysis for Models")
plt.show()
```
